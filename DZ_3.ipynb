- Отфильтруйте df из примера по условию: Age > 30 и Score >= 90.
- Создайте DataFrame с покупками (столбцы: User , Item , Quantity ), затем посчитайте общее

````python
import pandas as pd
import numpy as np
````

````python
df = pd.DataFrame({
    'Name': ['Alice','Bob','Charlie','David','Eva'],
    'Age': [25, 30, 35, 40, 22],
    'Score': [88, 92, 85, 90, 95]
})

df2 = pd.DataFrame({
    'Dept': ['HR','HR','IT','IT','HR'],
    'Salary': [100, 150, 200, 250, 180]
})

print("Исходные данные:")
print(df)
print("\ndf2:")
print(df2)
````

````python
# 1. Фильтрация: Age > 30 и Score >= 90
print("1. Фильтрация (Age > 30 и Score >= 90):")
filtered = df[(df['Age'] > 30) & (df['Score'] >= 90)]
print(filtered)
````

````python
print("2. Покупки и groupby():")
purchases = pd.DataFrame({
    'User': ['Alice', 'Bob', 'Alice', 'Charlie', 'Bob', 'Alice'],
    'Item': ['Apple', 'Banana', 'Orange', 'Apple', 'Orange', 'Banana'],
    'Quantity': [3, 2, 5, 1, 4, 2]
})
print("Данные о покупках:")
print(purchases)

total_by_user = purchases.groupby('User')['Quantity'].sum()
print("\nОбщее количество покупок по пользователям:")
print(total_by_user)
````

````python
# 3. Сортировка по нескольким столбцам
print("3. Сортировка df2 по Dept, потом по Salary:")
sorted_df2 = df2.sort_values(by=['Dept', 'Salary'])
print(sorted_df2)
````

````python
print("4. Средний возраст по классам:")
students = pd.DataFrame({
    'Name': ['Анна', 'Иван', 'Мария', 'Петр', 'Ольга', 'Сергей'],
    'Age': [15, 16, 15, 17, 16, 15],
    'Class': ['10A', '10B', '10A', '10B', '10A', '10B']
})
print("Данные о студентах:")
print(students)

avg_age_by_class = students.groupby('Class')['Age'].mean()
print("\nСредний возраст по классам:")
print(avg_age_by_class)
````
